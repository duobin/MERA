```python
import pandas as pd
import numpy as np
import os
```

```python
path = "../audio/spectrograms/"
```

```python
data = pd.read_csv("train.csv")
data.head()
```

```python
data.shape
```

```python
# get the names of all songs that have been downloaded
# load data for only those songs
songs = os.listdir(path)
songs = [np.int64(_[:-4]) for _ in songs]
```

```python
data = data[data["dzr_sng_id"].isin(songs)]
data.head()
```

```python
data
```

```python
# add the path
PATH = []
for i in range(data.shape[0]):
    image_path = os.path.join(path, str(data['dzr_sng_id'][i]) + '.png')
    PATH.append(image_path)
```

```python
data["Path"] = PATH
```

```python
data.head()
```

```python
data.drop("dzr_sng_id", axis=1, inplace=True)
```

```python
data.to_csv("mvp_data.csv", index=False)
```
